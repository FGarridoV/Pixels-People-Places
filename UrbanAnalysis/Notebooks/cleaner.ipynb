```python
import pandas as pd
import shutil
import os

df = pd.read_csv('zone_lists/phd_list.csv')

cond1 = (df['executed'] == 1)
cond2 = (df['processed'] == -1)

df = df.loc[cond1 & cond2]

folder_format = lambda city: f'zones/{city.capitalize()}_NL/'

for i,v in df.iterrows():
    folder = folder_format(v['name']) + "/detections"
    filetxt = folder_format(v['name']) + "/log_vp.txt"
    if os.path.exists(folder):
        shutil.rmtree(folder)
    if os.path.exists(filetxt):
        os.remove(filetxt)

    
```
