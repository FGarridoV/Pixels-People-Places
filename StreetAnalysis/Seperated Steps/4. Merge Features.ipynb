```python
import pickle
import numpy as np
```

```python
project_name = 'Delft'
image_features_folder = r'Data\Delft\image_features'
network_data_path = r'C:\Users\lange\OneDrive\Documenten\Jaar_5-Msc_CE_TTE\DSAIE_STREETS\Data_old\Delft_NL'

name = 'edges_drive'
```

```python
with open(r'Data\Delft\edges_drive_image_assignments.pkl', 'rb') as file:
    image_assignments = pickle.load(file)
```

```python
def create_section_vectors(image_assignments: dict[str, list[str]], image_features_folder: str) -> dict[str, np.ndarray]:
    section_vectors = {}

    count = 0
    for section, images in image_assignments.items():
        front_arrays = []
        back_arrays = []

        for image in images:
            array = np.load(f"{image_features_folder}\\{image}.npy")

            if image.endswith('_f.png'):
                front_arrays.append(array)
            else:  # image.endswith('_b.png'):
                back_arrays.append(array)

        front_array = np.mean(np.array(front_arrays), axis=0)
        back_array = np.mean(np.array(back_arrays), axis=0)
        
        section_vectors[section] = np.concatenate((front_array, back_array), axis=0)
        count += 1
        if count % 50 == 0:
            print(f"Section {count} out of {len(image_assignments)}")
    
    return section_vectors
```

```python
network_vectors = create_section_vectors(image_assignments, image_features_folder)
```

```python
with open(f"Data\\{project_name}\\{name}_vector_assignments.pkl", 'wb') as file:
    pickle.dump(network_vectors, file)
```
