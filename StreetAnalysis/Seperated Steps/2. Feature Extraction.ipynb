```python
import os
import multiprocessing

from extract_image_features import extract_features_to_file
```

```python
project_name = 'Delft'
images_folder = r"Z:\data\Delft_NL\imagedb"
```

Collect all front and back images from image folder

```python
image_filenames = []

existing_files = []
for file in os.listdir(r'C:\Users\lange\OneDrive\Documenten\Jaar_5-Msc_CE_TTE\DSAIE_STREETS\Scripts\Data\Delft\image_features'):
    existing_files.append(file[:-4])

for file in os.listdir(images_folder):
    if not file.endswith('_s_a.png') and not file.endswith('_s_b.png'):
        if file not in existing_files:
            image_filenames.append(file)
```

```python
# Number of images after filtering
print("Number of images: ", len(image_filenames))
```

```python
def extract_features(name: str, image_names: list[str], image_folder: str) -> None:
    """Input model and returns time it takes to create the feature extraction vectors"""

    # loop through each image in the dataset
    print(f"Starting images feature vector extraction for {name}: {len(image_filenames)} images given")
    print(f"Images will be taken from '{image_folder}' and saved to Data\\{name}\\image_features in the local workspace")

    if not os.path.exists('Data'):
        os.makedirs('Data')

    if not os.path.exists(f"Data\\{name}"):
        os.makedirs(f"Data\\{name}")

    if not os.path.exists(f"Data\\{name}\\image_features"):
        os.makedirs(f"Data\\{name}\\image_features")
    
    pool = multiprocessing.Pool() 
    pool.map(extract_features_to_file, image_names)
```

```python
extract_features(project_name, image_filenames, images_folder)
```
